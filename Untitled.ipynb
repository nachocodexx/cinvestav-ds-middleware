```python
import pandas as pd
import numpy as np
import os
from binascii import hexlify,unhexlify
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
sns.set(rc={'figure.figsize':(11.7,8.27)})

mms = MinMaxScaler()
```

```python
folders  = os.listdir('results')
data     = list(map(lambda x:x.split('.')[0],folders))
data     = list(map(lambda x: x.encode('utf-8'),data))
data     = list(map(lambda x: unhexlify(x),data))
data     = list(map(lambda x: str(x,'ascii'),data))
data     = list(map(lambda x: x.split(':'),data))
def toFloat2(x):
    fx = float
    return (list(map(fx,x[0])),list(map(fx,x[1]) ))
cols = ["SAMPLES","SIZE","INTER_ARRIVAL","READ_RATIO","SAS_SIZE","DISTRIBUTION","MEAN","STD","CONCURRENCY","AVG_IA","AVG_ST","DELAYS"]
df       = pd.DataFrame(data,columns=cols) 
data     = list(map(lambda x: (x[:9],x[9:]),data))
data     = list(map(toFloat2,data))
x = list(zip(data,folders))
y = list(map(lambda x: open("./results/"+x,'r'),folders))
y = list(map(lambda x: x.readlines(),y))
# len(y)
```

```python
def getQueueInf(x):
    x0 = list(map(lambda y:y.strip(),x))
    
    x1 = x0[-1]
    x2 = x1.split(',')
    # Bad solution,ik, butno time :c.
    x3 = list(map(lambda x: 0.0 if x =='r' else float(x),x2))
    return x2
w = y
# print(len(y))
z = list(map (getQueueInf,w) )
# print(len(z))
values =df.to_numpy().tolist()

values[0]+z[0]
v =list(zip(z,values))
v = list(map( lambda x:x[0]+x[1],v))
cols2 = ["AVG_DIQ","NIQ","SERVER_UTILIZATION","SIMULATION_TIME"]+cols
df2 = pd.DataFrame(v,columns=cols2)

df2
```
